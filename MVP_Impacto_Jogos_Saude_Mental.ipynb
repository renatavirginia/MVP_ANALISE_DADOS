#### 3.2.2.2 Verificação de Valores Discrepantes (Outliers)

```python
# Seleção das colunas numéricas para análise de outliers
colunas_numericas = df.select_dtypes(include=[np.number]).columns.tolist()

# Detecção de outliers via IQR
outliers_resumo = []
for col in colunas_numericas:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    n_outliers = ((df[col] < lower) | (df[col] > upper)).sum()
    if n_outliers > 0:
        outliers_resumo.append({
            'Variável': col,
            'Q1': round(Q1, 2),
            'Q3': round(Q3, 2),
            'IQR': round(IQR, 2),
            'Limite Inf.': round(lower, 2),
            'Limite Sup.': round(upper, 2),
            'Qtd Outliers': n_outliers,
            '% Outliers': round(n_outliers / len(df) * 100, 2)
        })

if outliers_resumo:
    df_outliers = pd.DataFrame(outliers_resumo)
    print(f'Variáveis com outliers detectados (método IQR):')
    display(df_outliers)
else:
    print('Nenhum outlier detectado pelo método IQR.')
```

**Principais achados:** A análise de outliers pelo método IQR (Interquartile Range) examinou todas as variáveis numéricas do dataset. A ausência de outliers extremos é consistente com o caráter sintético dos dados, onde os valores foram gerados dentro de faixas pré-definidas e com distribuições controladas.

**Ponto de atenção:** Em datasets reais, a presença de outliers é informativa , pode indicar jogadores com comportamento extremo (por exemplo, jogar mais de 16 horas por dia) ou erros de preenchimento. A decisão de remover ou manter outliers deve ser contextual e baseada no conhecimento do domínio, não apenas em critérios estatísticos.

## 3.2.6 Gráficos de Dispersão

Gráficos de dispersão permitem visualizar a relação entre duas variáveis numéricas, identificando possíveis correlações lineares ou não lineares.

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Amostra menor para scatter plots (performance)
df_scatter = df.sample(n=5000, random_state=42)

scatter_vars = [
    ('horas_diarias_jogo', 'escore_ansiedade', 'Horas de Jogo vs Ansiedade'),
    ('horas_diarias_jogo', 'escore_depressao', 'Horas de Jogo vs Depressão'),
    ('proporcao_jogos_violentos', 'escore_agressividade', 'Proporção Jogos Violentos vs Agressividade'),
    ('horas_exercicio', 'nivel_estresse', 'Horas de Exercício vs Estresse')
]

cores_scatter = [COR_ALERTA, COR_PRIMARIA, '#8E44AD', COR_TERCIARIA]

for ax, (x_var, y_var, titulo), cor in zip(axes.flatten(), scatter_vars, cores_scatter):
    ax.scatter(df_scatter[x_var], df_scatter[y_var], alpha=0.3, s=10, color=cor)
    
    # Linha de tendência
    z = np.polyfit(df_scatter[x_var].dropna(), df_scatter[y_var].dropna(), 1)
    p = np.poly1d(z)
    x_sorted = np.sort(df_scatter[x_var].dropna())
    ax.plot(x_sorted, p(x_sorted), color='black', linewidth=2, linestyle='--', label='Tendência')
    
    # Correlação
    corr = df_scatter[x_var].corr(df_scatter[y_var])
    ax.set_title(f'{titulo}\n(r = {corr:.3f})', fontsize=12, fontweight='bold')
    ax.set_xlabel(x_var.replace('_', ' ').title(), fontsize=10)
    ax.set_ylabel(y_var.replace('_', ' ').title(), fontsize=10)
    ax.legend(fontsize=9)
    ax.grid(alpha=0.3)

plt.suptitle('Gráficos de Dispersão - Relações Chave', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

**Principais achados:** Os gráficos de dispersão mostram a relação entre variáveis-chave. O coeficiente de correlação (r) indica a força e direção da relação linear. Valores de |r| próximos de 0 sugerem ausência de correlação linear, enquanto valores próximos de 1 indicam relação forte. As linhas de tendência (regressão linear) permitem visualizar a direção da relação.

**Ponto de atenção:** Correlações fracas (|r| < 0.3) entre horas de jogo e indicadores de saúde mental podem indicar que: (1) a relação não é linear, (2) existem variáveis confundidoras não controladas, ou (3) o dataset sintético não captura adequadamente esta relação. Em estudos reais, correlações moderadas (r ≈ 0.3-0.5) entre tempo de jogo e bem-estar são frequentemente reportadas.

## 4.1 Divisão Treino e Teste

```python
# Seleção de features numéricas para modelagem
features_modelagem = [
    'idade', 'horas_diarias_jogo', 'sessoes_semanais', 'anos_jogando',
    'horas_sono', 'horas_exercicio', 'consumo_cafeina',
    'nivel_estresse', 'escore_ansiedade', 'escore_depressao',
    'nivel_vicio', 'escore_solidao', 'escore_agressividade', 'escore_felicidade',
    'escore_interacao_social', 'proporcao_jogos_violentos',
    'proporcao_multiplayer', 'proporcao_jogos_noturnos',
    'horas_jogo_fds', 'desempenho_academico', 'produtividade_trabalho'
]

X = df[features_modelagem].dropna()

print(f'Features selecionadas: {len(features_modelagem)}')
print(f'Total de amostras: {X.shape[0]:,}')
```

A divisão treino/teste (80/20) garante que os modelos sejam avaliados em dados não vistos durante o treinamento, validando sua capacidade de generalização. O parâmetro `random_state=42` assegura **reprodutibilidade** , qualquer pessoa que execute o notebook obterá a mesma divisão.

**Ponto de atenção:** Para datasets muito grandes como este, a proporção 80/20 é adequada, pois mesmo 20% dos dados representa uma amostra significativa para teste (~20 mil registros).

## 4.2 Padronização, Tratamento com K-Means e PCA

```python
# Padronização com StandardScaler
scaler = StandardScaler()
```

```python
# Fit no treino, transform em ambos
X_treino_scaled = scaler.fit_transform(X_treino)
X_teste_scaled = scaler.transform(X_teste)

print('Padronização realizada com sucesso.')
print(f'\nMédia das features (treino padronizado): ~{X_treino_scaled.mean():.6f}')
print(f'Desvio padrão das features (treino padronizado): ~{X_treino_scaled.std():.6f}')
```

A padronização é realizada **apenas com base no conjunto de treino** para evitar vazamento de dados (data leakage). O `fit_transform` é aplicado ao treino e o `transform` ao teste.

```python
# Visualização dos dados padronizados
df_scaled = pd.DataFrame(X_treino_scaled, columns=features_modelagem)
df_scaled.head()
```

```python
# Histograma de uma variável padronizada
fig, ax = plt.subplots(figsize=(10, 5))

ax.hist(df_scaled['horas_diarias_jogo'], bins=30, color=COR_PRIMARIA, 
        edgecolor='white', alpha=0.8, density=True)
df_scaled['horas_diarias_jogo'].plot(kind='kde', ax=ax, color=COR_ALERTA, linewidth=2)

ax.set_title('Distribuição Padronizada - Horas Diárias de Jogo', fontsize=13, fontweight='bold')
ax.set_xlabel('Valor Padronizado (Z-score)', fontsize=12)
ax.set_ylabel('Densidade', fontsize=12)
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

**Principais achados:** Após a padronização, a variável `horas_diarias_jogo` apresenta distribuição centrada em zero com desvio padrão unitário, confirmando que o StandardScaler foi aplicado corretamente. A forma da distribuição (simetria, curtose) é preservada , apenas a escala muda.

**Ponto de atenção:** A padronização é realizada **apenas com base no conjunto de treino** (`fit_transform` no treino, `transform` no teste) para evitar vazamento de dados (data leakage). Este é um passo crítico que garante que informações do conjunto de teste não influenciem o pré-processamento.

```python
# Agrupamento com K-Means (3 clusters)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X_treino_scaled)
```

```python
# Centroides dos clusters
centroides = pd.DataFrame(kmeans.cluster_centers_, columns=features_modelagem)
centroides.index.name = 'Cluster'
print('Centroides dos Clusters:')
centroides.T
```

O algoritmo **K-Means** agrupa os dados em K clusters baseando-se na minimização da distância euclidiana intra-cluster. Cada centroide representa o "perfil médio" de jogador de cada cluster. A escolha de K=3 permite identificar perfis como: jogadores casuais, moderados e hardcore.

**Ponto de atenção:** O número de clusters (K=3) foi escolhido empiricamente. Idealmente, métodos como o **Elbow Method** ou o **Silhouette Analysis** com múltiplos valores de K deveriam ser utilizados para determinar o número ótimo de clusters.

```python
# Atribuição de clusters
clusters_treino = kmeans.predict(X_treino_scaled)
clusters_teste = kmeans.predict(X_teste_scaled)

print('Distribuição dos clusters (Treino):')
print(pd.Series(clusters_treino).value_counts().sort_index())

print(f'\nDistribuição dos clusters (Teste):')
print(pd.Series(clusters_teste).value_counts().sort_index())
```

```python
# Avaliação com Silhouette Score
sil_treino = silhouette_score(X_treino_scaled, clusters_treino)
sil_teste = silhouette_score(X_teste_scaled, clusters_teste)

print(f'Silhouette Score (Treino): {sil_treino:.4f}')
print(f'Silhouette Score (Teste): {sil_teste:.4f}')
```

**Principais achados:** O **Silhouette Score** mede a qualidade do agrupamento, variando de -1 a 1. Valores próximos de 1 indicam clusters bem definidos e separados. Valores próximos de 0 indicam sobreposição significativa entre clusters. Valores negativos indicam atribuição incorreta de amostras aos clusters.

**Ponto de atenção:** Um Silhouette Score baixo (< 0.25) sugere que os clusters não estão bem separados, o que é esperado quando as variáveis possuem correlações fracas (caso deste dataset sintético). Scores similares entre treino e teste indicam boa generalização do modelo de agrupamento.

```python
# Redução de dimensionalidade com PCA para visualização
pca = PCA(n_components=2, random_state=42)
X_treino_pca = pca.fit_transform(X_treino_scaled)
X_teste_pca = pca.transform(X_teste_scaled)
```

O **PCA (Principal Component Analysis)** reduz a dimensionalidade dos dados preservando a máxima variância, permitindo visualizar clusters em 2 dimensões.

```python
# Visualização dos clusters em 2D (PCA)
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Treino
scatter_treino = axes[0].scatter(
    X_treino_pca[:, 0], X_treino_pca[:, 1], 
    c=clusters_treino, cmap='viridis', alpha=0.3, s=5
)
axes[0].set_title(f'Clusters K-Means (Treino)\nSilhouette: {sil_treino:.4f}', 
                  fontsize=12, fontweight='bold')
axes[0].set_xlabel('Componente Principal 1', fontsize=10)
axes[0].set_ylabel('Componente Principal 2', fontsize=10)
plt.colorbar(scatter_treino, ax=axes[0], label='Cluster')

# Teste
scatter_teste = axes[1].scatter(
    X_teste_pca[:, 0], X_teste_pca[:, 1], 
    c=clusters_teste, cmap='viridis', alpha=0.3, s=5
)
axes[1].set_title(f'Clusters K-Means (Teste)\nSilhouette: {sil_teste:.4f}', 
                  fontsize=12, fontweight='bold')
axes[1].set_xlabel('Componente Principal 1', fontsize=10)
axes[1].set_ylabel('Componente Principal 2', fontsize=10)
plt.colorbar(scatter_teste, ax=axes[1], label='Cluster')

plt.suptitle('Visualização PCA dos Clusters K-Means', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print(f'\nVariância explicada pelo PCA: {pca.explained_variance_ratio_.sum():.2%}')
print(f'  - PC1: {pca.explained_variance_ratio_[0]:.2%}')
print(f'  - PC2: {pca.explained_variance_ratio_[1]:.2%}')
```

**Principais achados:** A visualização PCA em 2 dimensões mostra a projeção dos dados nos dois primeiros componentes principais. A variância explicada pelos dois componentes indica quanto da variabilidade total dos dados é capturada nesta projeção. Os clusters seguem distribuição similar entre treino e teste, indicando boa generalização.

**Ponto de atenção:** Se a variância explicada pelos dois componentes for baixa (< 50%), a visualização 2D pode não representar adequadamente a estrutura real dos dados em alta dimensão. Neste caso, os clusters podem parecer mais sobrepostos do que realmente são no espaço original.
